```python
# Import c√°c th∆∞ vi·ªán c·∫ßn thi·∫øt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import ast

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

# S·ª≠ d·ª•ng data ƒë√£ ƒë∆∞·ª£c x·ª≠ l√Ω

```python
# ƒê·ªçc file csv
data = pd.read_csv('../data/processed_data.csv', index_col=0)

# Chuy·ªÉn ƒë·ªïi object th√†nh list
data["do_phan_giai_cam_truoc"] = data["do_phan_giai_cam_truoc"].apply(ast.literal_eval)
data["do_phan_giai_cam_sau"] = data["do_phan_giai_cam_sau"].apply(ast.literal_eval)

data.info()
data.head()
```

- Gi√∫p x√°c ƒë·ªãnh v·ªã tr√≠ c·ªßa c√°c th∆∞∆°ng hi·ªáu tr√™n th·ªã tr∆∞·ªùng

## Ti·ªÅn x·ª≠ l√Ω

```python
#Kh√¥ng l·∫•y nh·ªØng ƒëi·ªán tho·∫°i c≈©
data_new = data[~data['ten'].str.contains('c≈©', na=False)]
df = data_new[["hang_dien_thoai", "gia_moi"]].copy().dropna()
df = df.groupby("hang_dien_thoai").agg(['mean', 'count']).sort_values(('gia_moi', 'mean'), ascending=False).reset_index()
df.columns = ['hang_dien_thoai','mean', 'count']
df = df[df['count']>10]
df
```

## Tr·ª±c quan h√≥a

```python
# Plotting the bar chart
plt.figure(figsize=(15, 6))
plt.bar(df["hang_dien_thoai"], df["mean"], color='skyblue')
plt.title("Gi√° ƒëi·ªán tho·∫°i trung b√¨nh theo d√≤ng", fontsize=14)
plt.xlabel("D√≤ng ƒëi·ªán tho·∫°i", fontsize=12)
plt.ylabel("Gi√° trung b√¨nh", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()
```

## Ti·ªÅn x·ª≠ l√Ω d·ªØ 

```python
df = data
df['chenh_lech_gia'] = df['gia_moi'] - df['gia_cu']
res = df.groupby(['hang_dien_thoai', 'la_dien_thoai_cu'])['gia_moi'].mean().unstack().reset_index()
res.columns = ['hang_dien_thoai', 'gia_moi_trung_binh', 'gia_cu_trung_binh']
res = res.dropna()
res
```

### S·ª± ch√™nh l·ªách c·ªßa gi√° c≈© v√† gi√° m·ªõi

```python
x = np.arange(len(res)) * (1 + 0.2)
width = 0.5 # ƒë√¢y l√† ƒë·ªô r·ªông c·ªßa t·ª´ng c·ªôt
plt.bar(x - width/2, res['gia_moi_trung_binh'], width , label = 'Gi√° ƒëi·ªán tho·∫°i m·ªõi', color = 'blue')
plt.bar(x + width/2, res['gia_cu_trung_binh'], width , label = 'Gi√° ƒëi·ªán tho·∫°i c≈©', color = 'orange')
plt.title('Gi√° trung b√¨nh ƒëi·ªán tho·∫°i c≈© v√† m·ªõi')
plt.xlabel('Gi√° (VNƒê)')
plt.ylabel('H√£ng ƒëi·ªán tho·∫°i')
plt.xticks(x, res['hang_dien_thoai'], rotation = 45)
plt.legend()
plt.show()
```

## Ph·∫ßn trƒÉm gi√° b·ªã m·∫•t ƒëi 

```python
res['lose_value'] = (res['gia_moi_trung_binh'] - res['gia_cu_trung_binh']) / res['gia_moi_trung_binh'] * 100
res = res.sort_values(by = 'lose_value')
res
```

```python
# visualize 
plt.bar(res['hang_dien_thoai'], res['lose_value'])
plt.ylabel('Ch√™nh l·ªách gi√° gi·ªØa m·ªõi v√† c≈© (%)')
plt.title('Bi·ªÉu ƒë·ªì th·ªÉ hi·ªán s·ª± ch√™nh l·ªách gi√° tr·ªã gi·ªØa ƒëi·ªán tho·∫°i m·ªõi v√† c≈©')
plt.xlabel('Th∆∞∆°ng hi·ªáu')
plt.xticks(rotation =45)
plt.show()
```
